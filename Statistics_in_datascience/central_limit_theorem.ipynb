```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python

# create random population 
np.random.seed(42)
population = np.random.uniform(10, 10000, 100000) #from 10 to 10000 of total 1lakh

# separate 1000 samples of 50 data
num_samples = 1000
sample_size = 50

sample_means = []

for i in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(sample.sum()/sample_size)

#Calculate average of the sample means
average_of_samples = sum(sample_means)/num_samples

# Print results
# print("Sample Means:", np.float64(sample_means))
print("Average of Sample Means:", average_of_samples)

#Plot histogram of sample means
plt.figure(figsize=(10,5))
count, bins, ignored = plt.hist(sample_means, bins=10, density=True, alpha=0.6)

#Fit a normal distribution curve
mu = sum(sample_means) / len(sample_means)
sigma = (sum((x-mu)**2 for x in sample_means) / len(sample_means))**0.5
x = np.linspace(min(bins), max(bins), 100)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y, linewidth=2)
plt.title("Distribution of Sample Means (Central Limit Theorem)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

plt.show()
```
