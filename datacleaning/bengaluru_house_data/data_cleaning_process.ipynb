```python
def list_diff_values_format(column):
    # 1. Get ALL unique values from the column
    all_unique_values = df[column].dropna().unique()
    unique_formats = {}
    for item in all_unique_values:
        item_str = str(item).strip() # remove leading/trailing whitespaces
        # 2. Categorize the format
        if '-' in item_str:
            format_type = 'Range'
        else:
            # Extract only alphabetic characters to find the unit
            unit = re.sub(r'[^a-zA-Z]', '', item_str)

            if unit == '': # if no unit, it's a number
                format_type = 'Number'
            else:
                format_type = unit # e.g., 'SqMeter', 'Acres'

        # 3. Store the first example of each format found
        if format_type not in unique_formats:
            unique_formats[format_type] = item_str

    # 4. Display the results
    print(f"{'FORMAT TYPE':<15} | {'EXAMPLE'}")
    print("-" * 35)
    for fmt, example in unique_formats.items():
        print(f"{fmt:<15} | {example}")
```

```python
# extract the number of bedrooms from the size column and create a new column called bhk
df['bhk'] = df['size'].str.extract('(\d+)').astype(int)

# smart imputation for the bath column using the median value of the bath column for each bhk
df['bath'] = df['bath'].fillna(
    df.groupby('bhk')['bath'].transform('median')
)

# round the bath column to the nearest integer and convert it to int type
df['bath'] = df['bath'].round().astype(int)
```

```python
# fill missing balcony as no balcony
df['balcony'] = df['balcony'].fillna(0)
```

```python
def convert_sqft_to_num(x):
    if isinstance(x, float):
        return x
    x = str(x).strip()
    # ---------- RANGE ----------
    if '-' in x:
        tokens = x.split('-')
        if len(tokens) == 2:
            try:
                return (float(tokens[0]) + float(tokens[1])) / 2
            except:
                return np.nan

    # ---------- NORMAL NUMBER ----------
    try:
        return float(x)
    except:
        pass

    # ---------- UNIT CONVERSIONS ----------
    unit_mapping = {
        'Sq. Meter': 10.7639,
        'Sq. Yards': 9,
        'Perch': 272.25,
        'Acres': 43560,
        'Cents': 435.6,
        'Guntha': 1089,
        'Grounds': 2400
    }
    for unit in unit_mapping:
        if unit in x:
            try:
                value = float(x.replace(unit, '').strip())
                return value * unit_mapping[unit]
            except:
                return np.nan
    return np.nan
```

```python
# Convert all values into numeric
df['total_sqft'] = df['total_sqft'].apply(convert_sqft_to_num)
```

```python
# Create useful columns
df['total_price'] = df['price'] * 100000
df['price_per_sqft'] = (df['total_price'] / df['total_sqft']).round(2)
```

```python
# now remove 'bhk' and 'price' column that are created during cleaning process
df.drop('bhk', axis=1, inplace=True)
df.drop('price', axis=1, inplace=True)
```

```python
# fix datatypes
cols = ['area_type', 'availability', 'location', 'size', 'society']
df[cols] = df[cols].astype(str).apply(lambda x: x.str.lower())
df['bath']=df['bath'].astype(int)
df['balcony']=df['balcony'].astype(int)
df['total_price']=df['total_price'].astype(int)

# reset the index of the dataframe
df.reset_index(inplace=True, drop=True)
```
