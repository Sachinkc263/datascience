```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ACLBSL_subset.csv") #csv file lai load garim
df.head(3)
```

```python
#We are working on single stock data so the symbol is only one.
stock_symbol = df["symbol"].unique()[0]
print(stock_symbol) # symbol of stock
```

1. Top Volume Days

```python
# 'volume' column lai descending order ma sort garxa, take first 5 rows with (date, symbol, volume)
top_volumes = df.sort_values(by= "volume", ascending=False).head(5)[["date", "symbol", "volume"]]
# index lai feri 1 bata start garxa
top_volumes.index = range(1,len(top_volumes)+1)
print("\nTop Volume Days:\n")
print(top_volumes)
```

```python
plt.figure()
#bar graph of top volume day
plt.bar(top_volumes["date"].astype(str), top_volumes["volume"])
plt.title(f"Top Volume Days ({stock_symbol})")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Note: The highest Volume of ACLBSL is on 2021/01/18 which is around 50000 kitta.

2. Lowest Volume Days 

```python
# "volume" column lai ascending order ma sort garxa, take 5 rows with ("date","symbol","volume")
low_volume = df.sort_values(by= "volume", ascending=True).head(5)[["date","symbol","volume"]]
#index lai feri 1 dekhi strat garinxa
low_volume.index = range(1, len(low_volume)+1)
print("\nLow Volume Days:\n")
print(low_volume)
```

```python
plt.figure()
#bar graph of lowes volume days
plt.bar(low_volume["date"].astype(str), low_volume["volume"])
plt.title(f"low Volume Days ({stock_symbol})")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Note: The lowest volume of ACLBSL is on 2025/12/24 which is 13 kitta only.

3. Top Volatile Days

```python
# volatility (price fluctuation) formula
df["volatility %"] = ((df['high']-df['low'])/df['close'])*100
#volatility % lai descending sort gari, take 5 rows with ('date','symbol','close', 'high', 'low', 'volatility %')
top_volatile = df.sort_values(by="volatility %", ascending=False).head(5)[['date','symbol','close', 'high', 'low', 'volatility %']]
# reindex from 1
top_volatile.index = range(1, len(top_volatile)+1)
print("\nTop Most Volatile days (based on daily range %):\n")
print(top_volatile)
```

```python
plt.figure()
#bar graph of top volatile days
plt.bar(top_volatile["date"], top_volatile["volatility %"])
plt.title(f"Top Volatile Days ({stock_symbol})")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

4. Low Volatile Days

```python
#volatility % lai ascending sort gari, take 5 rows with ('date','symbol','close', 'high', 'low', 'volatility %')
low_volatile = df.sort_values(by="volatility %", ascending=True).head(10)[['date','symbol','close', 'high', 'low', 'volatility %']]
#reindex from 1
low_volatile.index = range(1, len(low_volatile)+1)
print("\nTop Most Volatile days (based on daily range %):\n")
print(low_volatile)
```

```python
plt.figure()
#bargraph of low valatile days
plt.bar(low_volatile["date"], low_volatile["volatility %"])
plt.title(f"Low Volatile Days ({stock_symbol})")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Note: ACLBSL price fluctuates high on 2023/02/26 and low on 2022/07/14

5. Comparision of Top gains/losses

```python
# ensure datetime + sorting
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

#chages calculation per day
df["changes%"] = (df["close"].pct_change() *100).round(2)
df.head()
```

```python
top_changes = df.sort_values(by="changes%", ascending=False).head(5)[['date','symbol','changes%']]
low_changes = df.sort_values(by="changes%", ascending=True).head(5)[['date','symbol','changes%']]
top_changes.index = low_changes.index = range(1, 6)
print("\nTop Changes:\n")
print(top_changes)
print("\nLow Changes:\n")
print(low_changes)
```

```python
plt.figure()

plt.bar(top_changes["date"].astype(str), top_changes["changes%"], label="Top Gains")
plt.bar(low_changes["date"].astype(str), low_changes["changes%"], label="Top Losses")

plt.title("Top 5 Gains vs Losses")
plt.xlabel("Date")
plt.ylabel("Change %")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

Here is the combined Gain/Loss Comparison chart

6. Relationship between Trading Volume and Volatility

```python
plt.figure()
plt.scatter(df["volume"], df["volatility %"])

plt.xlabel("Volume")
plt.ylabel("Volatility %")
plt.title("Volume vs Volatility")
plt.show()
```

Note: This scatter plot shows whether higher trading activity is associated with larger price fluctuations.

7. Price Trend Over Time

```python
plt.figure()
plt.plot(df["date"], df["close"])
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.title("Price Trend")

plt.show()
```

Note: This line chart shows the long-term movement of the stock price across trading dates.

8. Distribution of Daily Changes

```python
plt.figure()
plt.hist(df["changes%"].dropna(), bins=40)

plt.xlabel("Daily Change %")
plt.ylabel("Frequency")
plt.title("Distribution of Daily Returns")
plt.show()
```

Note: This histogram shows how daily returns are distributed, indicating typical gains/losses and extreme movements.
