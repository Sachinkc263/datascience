```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ACLBSL_subset.csv") #csv file lai load garim
df.head(3)
```

```python
#We are working on single stock data so the symbol is only one.
stock_symbol = df["symbol"].unique()[0]
print(stock_symbol)
```

1. Top Volume

```python
top_volumes = df.sort_values(by= "volume", ascending=False).head(10)[["date", "symbol", "volume"]]
top_volumes.index = range(1,len(top_volumes)+1)
print("\nTop Volume Days:\n")
print(top_volumes)
```

```python
plt.figure()
plt.bar(top_volumes["date"], top_volumes["volume"])
plt.title(f"Top Volume Days ({stock_symbol})")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

The highest Volume of ACLBSL is on 2021/01/18 which is around 50000 kitta.

2. Lowest Volume

```python
low_volume = df.sort_values(by= "volume", ascending=True).head(10)[["date","symbol","volume"]]
low_volume.index = range(1, len(low_volume)+1)
print("\nLow Volume Days:\n")
print(low_volume)
```

```python
plt.figure()
plt.bar(low_volume["date"], low_volume["volume"])
plt.title(f"low Volume Days ({stock_symbol})")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

The lowest volume of ACLBSL is on 2025/12/24 which is 13 kitta only.

3. Top Volatility

```python
df["Volatility %"] = ((df['high']-df['low'])/df['close'])*100
top_volatile = df.sort_values(by="Volatility %", ascending=False).head(10)[['date','symbol','close', 'high', 'low', 'Volatility %']]
top_volatile.index = range(1, len(top_volatile)+1)
print("\nTop Most Volatile days (based on daily range %):\n")
print(top_volatile)
```

```python
plt.figure()
plt.bar(top_volatile["date"], top_volatile["Volatility %"])
plt.title(f"Top Volatile Days ({stock_symbol})")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

4. Low Volatility

```python
low_volatile = df.sort_values(by="Volatility %", ascending=True).head(10)[['date','symbol','close', 'high', 'low', 'Volatility %']]
low_volatile.index = range(1, len(low_volatile)+1)
print("\nTop Most Volatile days (based on daily range %):\n")
print(low_volatile)
```

```python
plt.figure()
plt.bar(low_volatile["date"], low_volatile["Volatility %"])
plt.title(f"Low Volatile Days ({stock_symbol})")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

ACLBSL price fluctuates high on 2023/02/26 and low on 2022/07/14

```python
# ensure datetime + sorting
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

df["changes%"] = (df["close"].pct_change() *100).round(2)
df.head()
```

```python
def summarize(df, name):
    return {
        "Name":name,
        "n": len(df),
        "mean_volatility": df["volatility %"].mean().round(2),
        "mean_return": df["changes%"].mean().round(2),
        "mean_volume": df["volume"].mean()
    }
summarize(df, "Population")
```
