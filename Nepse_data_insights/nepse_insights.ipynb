```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel("/content/python_project2.xlsx")
df.head()
```

```python
top_gainer = df.sort_values(by="Diff %",ascending = False).head(10)[['Symbol','Close','Diff %']]
top_gainer.index = range(1,len(top_gainer)+1)
print("\nTop Gainer:\n")
print(top_gainer)
```

```python
plt.figure()
plt.bar(top_gainer["Symbol"], top_gainer["Diff %"])
plt.title("Top 10 Gainers")
plt.xlabel("Stock")
plt.ylabel("Diff %")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 losers
top_loser = df.sort_values(by="Diff %",ascending = True).head(10)[['Symbol','Close','Diff %']]
top_loser.index = range(1,len(top_loser)+1)
print("\nTop Loser:\n")
print(top_loser)
```

```python
plt.figure()
top_loser['Diff %']=top_loser['Diff %'].astype(str).str.replace("-","").astype(float)
plt.bar(top_loser["Symbol"], top_loser["Diff %"])
plt.title("Top 10 Losers")
plt.xlabel("Stock")
plt.ylabel("Diff %")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.bar(top_vol["Symbol"], top_vol["Vol"])
plt.title("Top 10 Stocks by Volume")
plt.xlabel("Stock")
plt.ylabel("Volume")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_turnover = df.sort_values(by="Turnover",ascending = False).head(10)[['Symbol','Turnover']]
top_turnover.index = range(1,len(top_vol)+1)
print("\nTop 10 Stocks by Turnover:\n")
print(top_turnover)
```

```python
plt.figure()
plt.bar(top_turnover["Symbol"], top_turnover["Turnover"])
plt.title("Top 10 Stocks by Turnover")
plt.xlabel("Stock")
plt.ylabel("Turnover")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_transaction = df.sort_values(by="Trans.",ascending = False).head(10)[['Symbol','Trans.']]
top_transaction.index = range(1,len(top_vol)+1)
print("\nTop 10 Stocks by Transaction:\n")
print(top_transaction)
```

```python
df["Volatility %"] = ((df['High']-df['Low'])/df['Close'])*100
top_volatile = df.sort_values(by="Volatility %", ascending=False).head(10)[['Symbol','Close', 'High', 'Low', 'Volatility %']]
top_volatile.index = range(1, len(top_volatile)+1)
print("\nTop 10 Most Volatile Stocks (based on daily range %):\n")
print(top_volatile)
```

```python
plt.figure()
plt.bar(top_volatile['Symbol'],top_volatile['Volatility %'])
# sns.barplot(x='Symbol', y='Volatility %', data=top_volatile)
plt.title('Top 10 Stocks by Volatility %')
plt.xlabel('Stock Symbol')
plt.ylabel('Volatility %')
plt.xticks(rotation=45)
plt.show()
```
